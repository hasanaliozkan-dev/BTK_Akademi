#Random Sampling

```python
data = pd.read_csv('Ads_CTR_Optimisation.csv')
```

```python
import random

N = 10000
d = 10 
sum = 0
choosen = []
for n in range(0,N):
  Add = random.randrange(d)
  choosen.append(Add)
  award = data.values[n,Add] # verilerdeki n. satır = 1 ise odul 1
  sum = sum + award
```

```python
plt.hist(choosen)
plt.show()
```

#Upper Confidence Bound

```python
import math
```

```python
N = 10000
d = 10 
sum = 0
awards = [0]*d
clicks = [0]*d
total= 0
choosen = []
for n in range(0,N):
  add = 0
  max_ucb=0
  for i in range(0,d):
    if clicks[i]>0:
      average = awards[i]/clicks[i]
      delta=math.sqrt((3/2)*(math.log(n)/clicks[i]))
      ucb=average + delta
    else:
      ucb=N*10
    if max_ucb<ucb:
      max_ucb= ucb
      add=i
  clicks[add] = clicks[add]+1
  choosen.append(add)
  award = data.values[n,add]
  awards[add] = awards[add] + award
  total += award
total
```

```python
plt.hist(choosen)
plt.show()
```

#Thompson

```python
N = 10000
d = 10 
sum = 0
total= 0
choosen = []
ones= [0]*d
zeros= [0]*d
for n in range(0,N):
  add = 0
  max_th=0
  for i in range(0,d):
    rand_beta = random.betavariate(ones[i]+1,zeros[i]+1)
    if rand_beta >max_th:
      max_th=rand_beta
      add = i
  
  choosen.append(add)
  award = data.values[n,add]
  if award == 1:
    ones[add] += 1
  else: 
    zeros[add] +=1
  total += award
total
```

```python
plt.hist(choosen)
plt.show()
```
