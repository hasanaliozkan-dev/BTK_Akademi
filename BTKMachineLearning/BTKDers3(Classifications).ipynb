#KNN

```python
data = pd.read_csv("veriler.csv")
indep= data.iloc[:,1:4]
dep = data.iloc[:,-1:]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(indep,dep,test_size=0.33,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(x_train)
X_test = sc.transform(x_test)
```

#SVM

```python
data = pd.read_csv("veriler.csv")
indep= data.iloc[:,1:4]
dep = data.iloc[:,-1:]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(indep,dep,test_size=0.33,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(x_train)
X_test = sc.transform(x_test)
```

```python
from sklearn.svm import SVC
model_svc= SVC(kernel="rbf").fit(x_train,y_train)
y_pred = model_svc.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_pred,y_test)
```

#Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
model_gnb = GaussianNB().fit(X_train,y_train)
y_pred = model_gnb.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_pred,y_test)
```

#Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
model_dtc = DecisionTreeClassifier(criterion="entropy").fit(X_train,y_train)
y_pred = model_dtc.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_pred,y_test)
```

#Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
model_rf = RandomForestClassifier(criterion="entropy",n_estimators=10).fit(X_train,y_train)
y_pred = model_rf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_pred,y_test)
```

```python
x= data.iloc[:,0:4]
y= data.iloc[:,-1:]
```

##Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
y_pred = model_lr.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_pred,y_test)
```

##KNN

```python
from sklearn.neighbors import KNeighborsClassifier
model_knn = KNeighborsClassifier(metric="minkowski",n_neighbors=3).fit(X_train,y_train)
```

```python
y_pred = model_knn.predict(X_test)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

##Naive Bayes
