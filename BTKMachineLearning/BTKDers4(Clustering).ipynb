```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#K-Means

```python
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv("musteriler.csv")
```

```python
data
```

```python
x = data.iloc[:,3:].values
x
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3,init="k-means++").fit(x)
```

```python
kmeans.cluster_centers_
```

```python
results = list()
for i in range(1,11):
  kmeans = KMeans(n_clusters=i,random_state=123,init="k-means++")
  kmeans.fit(x)
  results.append(kmeans.inertia_)
results
```

```python
plt.plot(range(1,11),results);
```

```python
kmeans = KMeans(n_clusters=4,random_state=123,init="k-means++")
```

```python
y_pred = kmeans.fit_predict(x)
```

```python
plt.scatter(x[y_pred==0,0],x[y_pred==0,1],s=100,c='red')
plt.scatter(x[y_pred==1,0],x[y_pred==1,1],s=100,c='green')
plt.scatter(x[y_pred==2,0],x[y_pred==2,1],s=100,c='blue')
plt.scatter(x[y_pred==3,0],x[y_pred==3,1],s=100,c='yellow');
plt.title("Kmeans")
```

#Hiyerarşik Kümeleme

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model_hc = AgglomerativeClustering( n_clusters=3,affinity='euclidean',linkage='ward')
```

```python
y_pred = model_hc.fit_predict(x)
```

```python
y_pred
```

```python
plt.scatter(x[y_pred==0,0],x[y_pred==0,1],s=100,c='red')
plt.scatter(x[y_pred==1,0],x[y_pred==1,1],s=100,c='green')
plt.scatter(x[y_pred==2,0],x[y_pred==2,1],s=100,c='blue');
```

```python
import scipy.cluster.hierarchy as sch
```

```python
dendrogram = sch.dendrogram(sch.linkage(x,method="ward"))
```
