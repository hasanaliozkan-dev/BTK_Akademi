```
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10,9,20)
y = x**3
z = x**2
fgr = plt.figure()
axs = fgr.add_axes([0.1,0.1,0.8,0.8])
axs.plot(x,y,"b")
axs.set_xlabel("X Axis")
axs.set_ylabel("Y Axis")
axs.set_title("Cube")
axs2 = fgr.add_axes([0.2,0.6,0.25,0.25])
axs2.plot(x,z,"r")
axs2.set_xlabel("X Axis")
axs2.set_ylabel("Y Axis")
axs2.set_title("Square")
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10,9,20)
y = x**3
z = x**2
fgr = plt.figure()
axs = fgr.add_axes([0,0,1,1])
axs.plot(x,y,label="Square")
axs.plot(x,z,label="Cube")
axs.legend(loc=4)
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10,9,20)
y = x**3
z = x**2
fig,axs = plt.subplots(nrows=2,ncols=1)
axs[0].plot(x,y) 
axs[0].set_title("Cube")
axs[1].plot(x,z) 
axs[1].set_title("Square")
plt.tight_layout()
fig.savefig("figure1.pdf")
plt.show()
```
