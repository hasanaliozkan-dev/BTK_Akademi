```
import matplotlib.pyplot as plt
year = [2011,2012,2013,2014,2015]
player1=[8,10,12,7,9]
player2=[7,12,5,15,21]
player3=[18,20,22,25,19]
#Stac Plat
plt.plot([],[],color="y",label="Player1")
plt.plot([],[],color="r",label="Player2")
plt.plot([],[],color="b",label="Player3")
plt.stackplot(year,player1,player2,player3,colors=["y","r","b"])
plt.title("YEAR VS SCORES")
plt.xlabel("YEARS")
plt.ylabel("SCORES")
plt.legend()
plt.show()
```

```
import matplotlib.pyplot as plt
goalTypes = "Penaltı","Serbest Vuruş","Kaleye Atılan Şut"
goals = [12,35,7]
clr = ["y","r","b"]
plt.pie(goals,labels=goalTypes,colors=clr,shadow=True,explode=(0.05,0.05,0.05),autopct="%1.1f%%")
plt.show()
```
