```
def naive_add(x,y):
  assert len(x.shape) == 2
  assert x.shape == y.shape
  x = x.copy()
  for i in range(x.shape[0]):
    for j in range(x.shape[1]):
      x[i,j] += y[i,j]
  return x 
  
```

```
x = np.array([[33, 87, 21, 10, 7],
             [30, 89, 7, 4, 3],
             [27, 93, 14, 7, 1]])
y = np.array([[33, 87, 21, 10, 7],
             [30, 89, 7, 4, 3],
             [27, 93, 14, 7, 1]])
z = naive_add(x,y)
z
```
