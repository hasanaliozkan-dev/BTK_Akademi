# Data Science

We can use the different `Python`'s modules and libraries on the collab in order to analyze and visualize data.

Let's visualize with `Numpy` and `Matplotlib`.


```
import numpy as np
from matplotlib import pyplot as plt
ys = 50 + np.random.randn(50)
x = [x for x in range(len(ys))]

plt.plot(x,ys,"*")
plt.fill_between(x,ys,45,where=(ys>45),facecolor="m",alpha=0.5)
plt.title("Example 1")
plt.show()
```
