```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
x = [1,2,3,4]
y = [1,4,9,16]
plt.plot(x,y,"o--r")#Marker-line-type,line colour
plt.axis([0,6,0,20])
plt.title("Grafik Başlığı")
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.show()
```

```

x = np.linspace(0,2,100)
plt.plot(x,x,label="Linear",color="red")
plt.plot(x,x**2,label="Quadratic",color="blue")
plt.plot(x,x*x*x,label="Cubic",color="green")
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Simple Plot",color="red")
plt.legend()
plt.show()
```

```

x = np.linspace(0,2,100)
fig,axs = plt.subplots(3)
axs[0].plot(x,x,color="red")
axs[0].set_title("Linear",color="red")
axs[1].plot(x,x*x,color="blue")
axs[1].set_title("Quadratic",color="red")
axs[2].plot(x,x*x*x,color="green")
axs[2].set_title("Cubic",color="red")
plt.tight_layout()
plt.show()
```

```

x = np.linspace(0,2,100)
fig,axs = plt.subplots(2,2)
fig.suptitle("Grafik Başlığı")
axs[0,0].plot(x,x,color="red")
axs[0,1].plot(x,x**2,color="black")
axs[1,0].plot(x,x**3,color="blue")
axs[1,1].plot(x,x**4,color="green")
plt.show()
```
